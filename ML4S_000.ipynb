Introduction:

*   Machine Learning is a subset of data science and it is the science of letting the machine figure out how a task is done. <br> In simple terms if we want the answer for 5 + 3 we would typically do the following <br> `a = 5, b = 3` <br> `print(a+b)` <br> and we get our 8. <br><br> In machine learning however, we do not tell the machine what to do. We let the machine figure out what to do. <br>To illustrate, we feed the following equation: <br> `a ? b = 8` <br> we want the machine to figure out what to do between a and b to get this 8 <br>we call variables [a, b, and 8] our ***predictors***. Meanwhile, we call ? our ***target variable*** that we want to obtain.








There are three types of machine learning Algorithms. An algorithm is considered ***supervised*** when you have a target variable. <br> Example: <br>

`inputs = [smart, pretty, evil, ugly]` <br>
`labels = [1, 1, 0, 0]` <br>

Here each input is mapped to a label marking whether its a positive or a negative word. Since labels exist, this problem is supervised. <br>

When the labels are not available, we consider the algorithm ***unsupervised***. Unsupervised problems are solved through ***clustering***. In other words, grouping inputs that look similar together. <br> Example: <br>

`inputs = [-2, -4, 7, 6, 5]` <br>

This would result two clusters where cluster1 = [-2,-4] and cluster2 = [7,6,5]. A cluster can then be given a label; e.g. Negative_nums = cluster1. <br>

A third and final type is semi-supervised algorithms. It is applied when the data is a mixture of both the previous types. It is usually applied on two stages where labels are used first to train a model before using the same model to inference the unlabled data and retrain. <br> Example: <br>


`Age = [1993, 1996, 2001, 2008, 1999, 1970, ...,1994, 1975]` <br>
`labels = [Y, null, Z, Z, null, X,..., Y, null]` <br>

Seperate labelled from unlabelled data: <br>

`Age_labelled = [1993, 2001, 2008, 1970, ..., 1994]` <br>
`labels_1 = [Y, Z, Z, X,..., Y]` <br>

`Age_unlabelled = [1996, 1999,..., 1975]` <br>

Train a model using the labelled data <br>
`model.train(Age_labelled, labels_1)` <br>
use the trained model to label the unlabelled years <br>
`new_labels = model.predict(Age_unlabelled)` <br>
add the new labels to the dataset and train again now using the full dataset. <br>
`Age = Age_labelled + new_labels` <br>
`labels = labels_1 + new_labels` <br>
`model.train(Age, labels)` <br>

Our main focus will be on supervised learning so don't worry if the others are not clear. <br> This is just some theory that you should know. Now lets start!
