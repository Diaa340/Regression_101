This scatter plot looks like an exponentitial function. Thats all we need. <br>
generally, there are common functions that are usually used for fitting. these functions are known as "activation functions". All we need to do is to pick one that seems to fit our curve and apply it.

<img src="https://www.researchgate.net/profile/Aaron-Stebner-2/publication/341310767/figure/fig7/AS:890211844255749@1589254451431/Common-activation-functions-in-artificial-neural-networks-NNs-that-introduce_W640.jpg" width="730" height="559">

for this example, softmax looks reasonably good but it is not easy to implement. Hence, I will use sigmoid.


```python
X = df['Year']
y = df['Value']
```

```python
from scipy.optimize import curve_fit #used to fit the function to the datapoints we have

def sigmoid(x, Beta_1, Beta_2): #from the internet
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y

X =X/max(X) #Min Max Scaler but hard coded
y =y/max(y) #Min Max Scaler but hard coded
```

```python
import matplotlib.pyplot as plot

popt, pcov = curve_fit(sigmoid, X, y) #curve_fit returns the best parameters (popt) and the curve (not needed)
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plot.figure(figsize=(8,5))
regression_line = sigmoid(x, *popt) #using a star here means return all params
plot.plot(X, y, 'ro', label='data')
plot.plot(x,regression_line, linewidth=3.0, label='fit')
plot.legend(loc='best')
plot.ylabel('GDP')
plot.xlabel('Year')
plot.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plot
from scipy.optimize import curve_fit #used to fit the function to the datapoints we have

df = pd.read_csv('/content/china_gdp.csv')

X = df['Year']
y = df['Value']

def sigmoid(x, a, b, c):
     y = a * x**2 + b * x + c
     return y

X =X/max(X) #Min Max Scaler but hard coded
y =y/max(y) #Min Max Scaler but hard coded


popt, pcov = curve_fit(sigmoid, X, y) #curve_fit returns the best parameters (popt) and the curve (not needed)
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plot.figure(figsize=(8,5))
regression_line = sigmoid(x, *popt) #using a star here means return all params
plot.plot(X, y, 'ro', label='data')
plot.plot(x,regression_line, linewidth=3.0, label='fit')
plot.legend(loc='best')
plot.ylabel('GDP')
plot.xlabel('Year')
plot.show()
```
